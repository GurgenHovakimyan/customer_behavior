# Machine Learning classification

```python
target = data['Segmentation']
X = data.drop(columns=['Segmentation', 'ID'])
X.head()
```

```python
X = pd.get_dummies(X, columns=["Gender", "Ever_Married", "Graduated", "Profession", "Spending_Score", "Var_1"])
```

```python
train_x, test_x, train_y, test_y = train_test_split(
    X, target, test_size=0.25, random_state=42)
print(
    f"Number of observations \n Train set: {len(train_x)}\n Test set: {len(test_x)}")
```

```python
train_x = StandardScaler().fit(train_x).transform(train_x)
```

```python
test_x = StandardScaler().fit(test_x).transform(test_x)
```

```python
def make_prediction(train_data = train_x, train_label = train_y, test_data = test_x, test_label = test_y, model = "Logistic",
 rand_st = 42, n_neighbors = 11, n_estimators=100, decision_function_shabe = 'ovr', kernel = 'rbf', standartize = False):
    if standartize:
        train_data = StandardScaler().fit(train_data).transform(train_data)
        test_data = StandardScaler().fit(test_data).transform(test_data)
    if model == "Logistic":
        Logistic = LogisticRegression(random_state=rand_st)
        Logistic.fit(train_data, train_label)
        prediction_y = Logistic.predict(test_data)
        cm = confusion_matrix(test_label, prediction_y)
    elif model == "Knn":
        Knn = KNeighborsClassifier(n_neighbors=n_neighbors)
        Knn.fit(train_data, train_label)
        prediction_y = Knn.predict(test_data)
        cm = confusion_matrix(test_label, prediction_y)
    elif model == "RFC":
        RFC = RandomForestClassifier(n_estimators=n_estimators, random_state=rand_st)
        RFC.fit(train_data, train_label)
        prediction_y = RFC.predict(test_data)
        cm = confusion_matrix(test_label, prediction_y)
    elif model == "SVM":
        suppor_vector_machine = SVC(decision_function_shape=decision_function_shabe, kernel = kernel)
        suppor_vector_machine.fit(train_data, train_label)
        prediction_y = suppor_vector_machine.predict(test_data)
        cm = confusion_matrix(test_label, prediction_y)

    fig, ax = plot_confusion_matrix(conf_mat=cm)
    fig.set_size_inches(18.5, 10.5, forward=True)
    ax.title.set_text(f"Confusion Matrix Prediction results of {model} model")
    print(f"Total accuracy: {(cm[0][0] + cm[1][1] + cm[2][2] + cm[3][3])/sum(sum(cm))}")
```

```python
make_prediction(model="Logistic")
```

```python
make_prediction(model="Knn", standartize = True)
```

```python
make_prediction(model="RFC", standartize = True)
```

```python
make_prediction(model = "SVM", standartize = True)
```

### AS we can see we have problem with some of labels: 
* Logistic Regression predicts 0, 2, 3 very well but label 1 is very terrible
* Knn makes model better but misclassification is very high 
* Random forest performance is better but still not the best
* Same for suppor vector machine

In the best model total accuracy is 48%. I don't need to look at TPR FPR etc because I have balanced data and from confusion matrix I can see the results without these calculations.

Also no information rate is 27.11% so our models are giving not bad results.

# Trying Boosting models

```python
import xgboost as xgb
```

```python
dtrain = xgb.DMatrix(data=train_x, label=train_y)
dtest = xgb.DMatrix(data=test_x, label=test_y)
```

```python
param = {
    'max_depth': 3,  # the maximum depth of each tree
    'eta': 0.01,  # the training step for each iteration
    'silent': 1,  # logging mode - quiet
    'objective': 'multi:softprob',  # error evaluation for multiclass training
    'num_class': 4,
    'gamma': 3}  # the number of classes that exist in this datset
num_round = 1000
```

```python
bst = xgb.train(param, dtrain, num_round)
```

```python
preds = bst.predict(dtest)
```

```python
preds
```

```python
best_preds = np.asarray([np.argmax(line) for line in preds])
```

```python
cm = confusion_matrix(test_y, best_preds)
```

```python
    fig, ax = plot_confusion_matrix(conf_mat=cm)
    fig.set_size_inches(18.5, 10.5, forward=True)
    ax.title.set_text("Confusion Matrix Prediction results of XGBoost model")
```

# Conclusion:

XGBoost algorithm gave us best result. Unfortunately our total accuracy is 53% which is not too high, but if we compare it with no information rate our model is significant, because it fits more than two times better than no information rate.
